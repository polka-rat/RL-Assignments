```python
def plot_analysis(logfile, verbose):
    with open(logfile, "rb") as f:
        logs = pickle.load(f)

    plt.figure(figsize=(6, 4))
    plt.plot(logs["episode"], logs["epsilon"])
    plt.xlabel("Episode")
    plt.ylabel("Epsilon")
    plt.title("Epsilon Decay")


    plt.tight_layout()
    if verbose:
        plt.show()

    return logs["episode_reward"], logs["avg_reward_200"], logs["episode"]
```
